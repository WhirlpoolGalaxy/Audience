```python
import time
import os
import segment_public_api
import csv
from segment_public_api.models.list_audiences200_response import ListAudiences200Response
from segment_public_api.models.pagination_input import PaginationInput
from segment_public_api.rest import ApiException
from pprint import pprint

# Configure Bearer authorization: token
configuration = segment_public_api.Configuration(
    access_token=BEARER_TOKEN
)

# Function to write data to CSV
def write_to_csv(audiences, filename="audience_data.csv"):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        header = ["id", "space_id", "name", "description", "key", "enabled", "query", "definition_type", "status", "created_by", "updated_by", "created_at", "updated_at", "include_historical_data", "include_anonymous_users"]
        writer.writerow(header)
        
        for audience in audiences:
            row = [
                audience.id,
                audience.space_id,
                audience.name,
                audience.description,
                audience.key,
                audience.enabled,
                audience.definition.query,
                audience.definition.type,
                audience.status,
                audience.created_by,
                audience.updated_by,
                audience.created_at,
                audience.updated_at,
                audience.options.include_historical_data,
                audience.options.include_anonymous_users
            ]
            writer.writerow(row)

# Enter a context with an instance of the API client
with segment_public_api.ApiClient(configuration) as api_client:
    # Create an instance of the API class
    api_instance = segment_public_api.AudiencesApi(api_client)
    space_id = space_id  # str
    pagination = segment_public_api.PaginationInput(count=10)  # PaginationInput

    try:
        # List Audiences
        api_response = api_instance.list_audiences(space_id, pagination=pagination)
        print("The response of AudiencesApi->list_audiences:\n")
        pprint(api_response)
        
        # Extract audience summaries
        audiences = api_response.data.audiences
        
        # Write data to CSV
        write_to_csv(audiences)
        print(f"Data successfully written to CSV.")
        
    except ApiException as e:
        print("Exception when calling AudiencesApi->list_audiences: %s\n" % e)
```
